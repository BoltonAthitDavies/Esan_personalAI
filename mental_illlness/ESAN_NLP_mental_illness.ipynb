# **Feature extraction**

```python
df.post.values[0]
```

```python
vectorizer = CountVectorizer()

# ให้ทำการแปลงข้อความให้เป็น Feature โดยการระบุข้อมูลที่ใช้ในการแปลงให้ถูกต้อง

X = vectorizer.fit_transform(df.post.values)
```

```python
len(X[0].toarray()[0])
```

```python
print(X.shape)
print(X.toarray())
print(vectorizer.get_feature_names_out())
```

# **Split Data into training and test sets**

```python
df['label'] = pd.Categorical(df.subreddit).codes
```

```python
df.subreddit
```

```python
df
```

```python
df['label'].unique()
```

```python
# ให้ระบุคอลัมน์ ที่จะใช้เป็น label เพื่อที่จะใช้ในการสอนและทำนาย

y = df.label.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

# **Modelling**

```python
log = LogisticRegression(random_state=0).fit(X_train, y_train)

print(log.predict(X_test[:10]))
print(log.predict_proba(X_test[:10]))
print(log.score(X_test, y_test))
```

**ให้ทดลองสร้างโมเดล ด้วย Algorithms อื่นๆ ด้วยตัวเอง อีก 2 โมเดล**

```python
from sklearn.linear_model import RidgeClassifier
clf1 = RidgeClassifier().fit(X_train, y_train)

print(clf1.predict(X_test[:10]))
# print(clf1.predict_proba(X_test[:10]))
print(clf1.score(X_test, y_test))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
from sklearn import svm
```

```python
X_test[0][0]
```

```python
clf2 = svm.SVC(kernel='rbf')
classifier = clf2.fit(X_train, y_train)

ypred_test = clf2.predict(X_test)
acc = format(accuracy_score(y_test, ypred_test), '0.2%')

labels = ['MentalHealth', 'Teaching']
# data = [[48000, 2, 25, 50000], [30000, 1, 22, 800000]]
predict_result = clf2.predict(X_test[:10])

i = 0
for result in predict_result:
  print(y_test[i], ':', labels[result])
  i += 1
print('acc:', acc)

disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, display_labels=labels)#, display_labels=labels)#, cmap=plt.cm.Blues)
print(disp.confusion_matrix)
print(classification_report(y_test, ypred_test, target_names=labels))
```

```python
pca = PCA(n_components=100, whiten=True) # n_components สามารถแก้ไขจำนวน Dimensions ตามต้องการได้

X = X
y = y


X = pca.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)

log = LogisticRegression(random_state=0).fit(X_train, y_train)

print(log.predict(X_test[:10]))
print(log.predict_proba(X_test[:10]))
print(log.score(X_test, y_test))
```

```python
# ระบุ y_true และ y_pred ด้วยตัวเองให้ถูกต้อง เพื่อใช้แสดงประสิทธิภาพของโมเดล


y_true = y_test
y_pred = log.predict(X_test)


target_names = ['MentalHealth', 'Teaching']
print(classification_report(y_true, y_pred, target_names=target_names))
```
